Un exemple pour dessiner des ronds dans un graphiques.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
x, y = [], []
for i in range(0, 3):
    for j in range(0, 4):
        x.append(i)
        y.append(j)
size = [300 for c in x]
colors = ["r" for c in x]
ax.scatter(x, y, s=size, c=colors, alpha=0.5);
```
